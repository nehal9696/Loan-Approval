```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('train_l.csv')
train.head()
```

```python
test = pd.read_csv('test_l.csv')
test.head()
```

```python
train.drop('Loan_ID',axis='columns',inplace=True)
```

```python
lid = test.Loan_ID
test.drop('Loan_ID',axis='columns')
```

```python
categorical_features = train.select_dtypes(include=['object']).columns
categorical_features
```

```python
le = LabelEncoder()
```

```python
cols = ( 'Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'Property_Area', 'Loan_Status')
```

```python
train_x = train
for c in cols:
    le.fit(list(train_x[c]))
    train_x[c]=le.transform(list(train_x[c]))
```

```python
train_x.head(4)
```

```python
train_x.columns[train_x.isna().any()]
```

```python
train_x['LoanAmount'] = train_x['LoanAmount'].fillna(train_x['LoanAmount'].mean())
train_x['Loan_Amount_Term'] = train_x['Loan_Amount_Term'].fillna(train_x['Loan_Amount_Term'].mean())
train_x['Credit_History'] = train_x['Credit_History'].fillna(train_x['Credit_History'].mean())
```

```python
y=train_x['Loan_Status']
train_x.drop('Loan_Status',axis='columns',inplace=True)
```

```python
categorical_features = test.select_dtypes(include=['object']).columns
categorical_features
```

```python
test_x = test.drop('Loan_ID',axis='columns')
cols1 = ('Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'Property_Area')
for c in cols1:
    le.fit(list(test_x[c]))
    test_x[c]=le.transform(list(test_x[c]))
```

```python
test_x.head()
```

```python
test_x.columns[test_x.isna().any()]
```

```python
test_x['LoanAmount'] = test_x['LoanAmount'].fillna(test_x['LoanAmount'].mean())
test_x['Loan_Amount_Term'] = test_x['Loan_Amount_Term'].fillna(test_x['Loan_Amount_Term'].mean())
test_x['Credit_History'] = test_x['Credit_History'].fillna(test_x['Credit_History'].mean())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(solver='liblinear',multi_class='auto',C=100)
```

```python
model.fit(x_train,y_train)
```

```python
  model.score(x_test,y_test)
```

```python
y_predict = model.predict(test_x)
```

```python
sub = pd.DataFrame()
```

```python
sub['Loan_ID'] = test.Loan_ID
sub['Loan_Status'] = y_predict
```
